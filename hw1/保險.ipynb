```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
import numpy as np
import codecs
with codecs.open('CLAIM.csv', 'rb', 'utf-8') as data:
    claim = csv.reader(data)
    sex=[]
    for row in claim:
            sex.append(row[1])




     
```

```python
with codecs.open('CLAIM.csv', 'rb', 'utf-8') as data:
    claim = csv.reader(data)
    bb=[]
    
    for row in claim:
            bb.append(row)
    bbb=np.array(bb)
    #print(bbb[:,16])
    #print(len(bb))
    rebuy=[]
    for i in range(len(bb)):
        if bb[i][16]!='':
            rebuy.append(bb[i]) #第i格row，的第17格有東西，表示有再購，加入rebuy
    #print(rebuy)
    
        
```

```python
rebuy_sex=[]
for row in rebuy:
    rebuy_sex.append(row[1])
#print(rebuy_sex)
```

```python
from collections import Counter
#matplotlib.pyplot.bar(left, height, alpha=1, width=0.8, color=, edgecolor=, label=, lw=3)
plt.figure(figsize=(9,15))

x=['M','F','rebuy-M','rebuy-F']
y=[sex.count('M'),sex.count('F'),rebuy_sex.count('M'),rebuy_sex.count('F')]
plt.bar(x,y,width=0.3 )

for a,b in zip(x,y):

    plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=7)
#為數字加上標籤
plt.xlabel('all/rebuy of female or male')
plt.ylabel('number')
plt.show()
```

```python
#從比例上看，不管是否再購都是女性比例較高
```

```python
LIFE=[]
ED=[]
ILP=[]
PA=[]
HEALTH=[]
DAILY=[]
CANCER=[]
ANNUITY=[]

#將再購的保單分類
for i in range(len(bb)):
        if bb[i][16]=='1.LIFE':
            LIFE.append(bb[i])
#print(len(LIFE))
for i in range(len(bb)):
        if bb[i][16]=='2.ED':
            ED.append(bb[i])
for i in range(len(bb)):
        if bb[i][16]=='3.ILP':
            ILP.append(bb[i])
for i in range(len(bb)):
        if bb[i][16]=='4.PA':
            PA.append(bb[i])
for i in range(len(bb)):
        if bb[i][16]=='5.HEALTH':
            HEALTH.append(bb[i])
for i in range(len(bb)):
        if bb[i][16]=='6.DAILY':
            DAILY.append(bb[i])
for i in range(len(bb)):
        if bb[i][16]=='7.CANCER':
            CANCER.append(bb[i])
for i in range(len(bb)):
        if bb[i][16]=='8.ANNUITY':
            ANNUITY.append(bb[i])
#len(保單類型)就是再保這種保單的數目
```

```python
fig = plt.figure()
labels = ["LIFE","ED","ILP","PA","HEALTH","DAILY","CANCER","ANNUITY"]
size = [len(LIFE), len(ED), len(ILP) , len(PA), len(HEALTH), len(DAILY),len(CANCER),len(ANNUITY)]
#explode = [0 , 0, .05, 0, 0, 0]  可設定圓餅是否突出，0則不突凸出，值越大 則凸出越大
plt.axis('equal')
plt.pie(size, labels = labels, autopct= "%1.1f%%")
# autopct-->項目百分比的格式
```
